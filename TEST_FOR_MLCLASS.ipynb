```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Fixing random state for reproducibility
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

axes[0] = plt.subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5)

axes[1] = plt.subplot(1,2,2)
plt.plot([0.2,0.5,0.8],[0.2,0.8,0.9])


for i in range(2): 
    axes[i].set_ylim(0,2)
    axes[i].spines['right'].set_visible(False) 
    axes[i].spines['top'].set_visible(False)  # no spine at the top
    axes[i].yaxis.set_ticks_position('left')  
    axes[i].spines['bottom'].set_visible(False)
    axes[i].yaxis.set_ticks_position('left')  
    axes[i].xaxis.set_ticks_position('bottom') 
    axes[i].spines['left'].set_position(('outward',15)) 
    axes[i].spines['bottom'].set_position(('outward',10)) 
    axes[i].spines['top'].set_position(('axes',5)) 
    axes[i].get_yaxis().set_tick_params(direction='in')
    axes[i].get_xaxis().set_tick_params(direction='in')
    axes[i].plot([-1,3], [1,1] , 'k--', color='#000000', linewidth=1)#gray line for Zero or 1        
    axes[i].tick_params(axis='both', which='major', labelsize=16)     
    axes[i].set_ylabel('some axis label', fontsize=14)
    axes[i].xaxis.set_visible(False)

    
plt.show()
fig.show()
```
